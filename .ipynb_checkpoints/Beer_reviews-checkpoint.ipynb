```python
import re

df['Style'] = df['Style'].str.lower()
df['Style'] = df['Style'].apply(lambda x: re.sub('style: ', '', x))
df
```

```python
df_filtered = df.groupby('Style').filter(lambda x: len(x) > 8)
df_filtered
```

```python
df_filtered['Brewer_comments'] = df_filtered['Brewer_comments'].str.lower()
df_filtered['Brewer_comments'] = df_filtered['Brewer_comments'].astype(str)
df_filtered['Brewer_comments'] = df_filtered['Brewer_comments'].apply(lambda x: re.sub('\s+', ' ', x))
df_filtered['Brewer_comments'] = df_filtered['Brewer_comments'].apply(lambda x: re.sub('[^\w\s]', '', x))
df_filtered
```

```python
df_filtered['Panel_comments'] = df_filtered['Panel_comments'].str.lower()
df_filtered['Panel_comments'] = df_filtered['Panel_comments'].astype(str)
df_filtered['Panel_comments'] = df_filtered['Panel_comments'].apply(lambda x: re.sub('\s+', ' ', x))
df_filtered['Panel_comments'] = df_filtered['Panel_comments'].apply(lambda x: re.sub('[^\w\s]', '', x))
df_filtered
```

```python
df_filtered['Editor_comments'] = df_filtered['Editor_comments'].str.lower()
df_filtered['Editor_comments'] = df_filtered['Editor_comments'].astype(str)
df_filtered['Editor_comments'] = df_filtered['Editor_comments'].apply(lambda x: re.sub('\s+', ' ', x))
df_filtered['Editor_comments'] = df_filtered['Editor_comments'].apply(lambda x: re.sub('[^\w\s]', '', x))
df_filtered
```
