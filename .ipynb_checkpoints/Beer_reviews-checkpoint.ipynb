```python
import nltk

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
stop
```

```python
df['Brewer_comments'] = df['Brewer_comments'].str.lower()
df['Brewer_comments'] = df['Brewer_comments'].astype(str)
df['Brewer_comments'] = df['Brewer_comments'].apply(lambda x: re.sub('\s+', ' ', x))
df['Brewer_comments'] = df['Brewer_comments'].apply(lambda x: re.sub('[^\w\s]', '', x))
df['Panel_comments'] = df['Panel_comments'].str.lower()
df['Panel_comments'] = df['Panel_comments'].astype(str)
df['Panel_comments'] = df['Panel_comments'].apply(lambda x: re.sub('\s+', ' ', x))
df['Panel_comments'] = df['Panel_comments'].apply(lambda x: re.sub('[^\w\s]', '', x))
df['Editor_comments'] = df['Editor_comments'].str.lower()
df['Editor_comments'] = df['Editor_comments'].astype(str)
df['Editor_comments'] = df['Editor_comments'].apply(lambda x: re.sub('\s+', ' ', x))
df['Editor_comments'] = df['Editor_comments'].apply(lambda x: re.sub('[^\w\s]', '', x))
df_filtered
```

```python
df_nlp = df.copy()
```

```python
df_nlp['Brewer_comments'] = df_nlp['Brewer_comments'].apply(lambda text: " ".join(word for word in text.split() if word not in stop))
df_nlp['Panel_comments'] = df_nlp['Panel_comments'].apply(lambda text: " ".join(word for word in text.split() if word not in stop))
df_nlp['Editor_comments'] = df_nlp['Editor_comments'].apply(lambda text: " ".join(word for word in text.split() if word not in stop))
df_nlp
```

```python
from nltk.tokenize import word_tokenize
#nltk.download('punkt')

df_nlp['bc_tokens'] = df_nlp.apply(lambda row: nltk.word_tokenize(row['Brewer_comments']), axis=1)
df_nlp['bc_tokens']
```

```python
brew_comment_tokens = df_nlp.bc_tokens.sum()
brew_comment_tokens
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

bc_tokens_stemmed = ' '.join([stemmer.stem(word) for word in brew_comment_tokens])
bc_tokens_stemmed
```

```python
from nltk import WordNetLemmatizer

lemztr = WordNetLemmatizer()
bc_token_stem_lem = lemztr.lemmatize(bc_tokens_stemmed)
bc_token_stem_lem
```

```python
from wordcloud import WordCloud

wc = WordCloud(background_color="white", max_words=2000, width=800, height=400)
# generate word cloud
wc.generate(bc_token_stem_lem)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
